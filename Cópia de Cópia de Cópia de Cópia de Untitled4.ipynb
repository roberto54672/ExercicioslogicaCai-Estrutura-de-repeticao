```python
# 14. Faça um programa que receba a idade de 15 pessoas
ate15 = 0
de16a30 = 0
de31a45 = 0
de46a60 = 0
maisde60 = 0
for pessoa in range(15):
  idade = int(input(f"digite a idade da pessoa{pessoa +1}"))
  if idade <= 15:
    ate15 += 1
  elif 16 <= idade <=30:
    de16a30 += 1
  elif 31 <= idade <= 45:
    de31a45 += 1
  elif 46 <= idade <= 60:
    de46a60 += 1
  else:
    maisde60 += 1

print(f"quantidade de pessoas por faixa etária\nAté 15 anos: {ate15} pessoas \nDe 16 até 30 anos:{de16a30} pessoas ")
```
